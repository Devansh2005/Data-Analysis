```python
# JOINT And RESID
```

```python
t=sns.load_dataset("tips")
```

```python
t.head()
```

```python
# #Methods
# RESID
# JOINT
```

```python
sns.jointplot(x="total_bill",y="tip",data=t)
```

```python
sns.residplot(x="total_bill",y="tip",data=t)
```

```python
sns.jointplot(x="total_bill",y="tip",data=t,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=t,kind="hex")
```

```python
#Regression plot
sns.jointplot(x="total_bill",y="tip",data=t,kind="reg")
```

```python
# Pair plot  -- kind of distribution plot( subplots)
```

```python
sns.pairplot(data=t, hue="sex")  # Hue for male and female
```

```python
#making these sub plots as distributed plots
x=sns.pairplot(data=t, hue="sex")  # Hue for male and female
x.map_diag(sns.distplot)
```

```python
x=sns.pairplot(data=t, hue="sex")  # Hue for male and female
x.map_upper(sns.kdeplot)
```
